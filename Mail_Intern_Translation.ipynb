```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# подгружаем модели
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-ru")
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-ru")
```

```python
def translate(sentence: str) -> str:
  """Принимает на вход строку на eng и выозвращает строку на ru"""
  global tokenizer
  global model
  translated_tokens = model.generate(**tokenizer(sentence, return_tensors='pt'))
  return tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
```

```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
from nltk.tokenize import sent_tokenize

def get_sentence() -> str:
  """Генератор, который возвращает предложения из датасета 20newsgroups"""
  X, _ = fetch_20newsgroups(return_X_y=True)
  for item in X:
    # разбиваем на абзацы и пропусаем первый (в нем мета-информация)
    for paragraph in item.split('\n\n')[1:]:
      for sentence in sent_tokenize(paragraph):
        yield sentence.replace('\n', '')

num_of_sentence_to_translate = 100
for i, sentence in enumerate(get_sentence()):

  print("\nSentence:", i + 1)
  print(sentence)
  print(translate(sentence))

  if i + 1 == num_of_sentence_to_translate:
    break
```
